# Исследование психического здоровья учащихся

Ссылка на Датасет: https://www.kaggle.com/datasets/sonia22222/students-mental-health-assessments

Набор данных представляет собой оценки психического здоровья студентов. Этот набор данных призван предоставить ценную информацию о психическом здоровье студентов, отражая ряд факторов, которые могут повлиять на их психическое здоровье.
Набор данных включает в себя богатую коллекцию записей, тщательно отобранных из различных анонимных источников.

Age: Возраст

Course: Курс (специальность)

Gender: Пол

CGPA: Кумулятивный средний балл

Stress_Level: Уровень стресса, который испытывает индивид

Depression_Score: Балл, отражающий уровень депрессии, испытываемой человеком

Anxiety_Score: Оценка, отражающая уровень тревожности, испытываемый индивидом

Sleep_Quality: Качество сна, испытываемого человеком

Physical_Activity: Уровень физической активности

Diet_Quality: Качество рациона питания человека

Social_Support: Уровень социальной поддержки, получаемой индивидом

Substance_Use: Частота употребления психоактивных веществ, таких как алкоголь, сигареты или другие наркотики

Family_History: Есть ли у данного человека в семейном анамнезе проблемы с психическим здоровьем

Chronic_Illness: Хронические заболевания

Counseling_Service_Use: Использования консультационных услуг

Financial_Stress: Уровень финансового стресса, испытываемого индивидом (между 0 и 5).

Extracurricular_Involvement: Внеклассная деятельность

Semester_Credit_Load: Количество зачетных единиц, которые студент получает в течение семестра (15-30)

**Задача:** Найти факторы, которые наиболее сильно влияют на ухудшение психического здоровья.

**Решение:** Описать портрет участника, наиболее подверженного депрессии.

Напишем функцию для построения круговых диаграмм с распределениям по уровням стресса

```python
def pie_by_stress(col_series):
  plt.figure(figsize=(8, 4))
  plt.suptitle(f"Распределение уровней стресса для {col_series.name}")

  num_of_unique = len(col_series.unique())

  for i, value in enumerate(sorted(col_series.unique())):
    ax = plt.subplot(1, num_of_unique, i+1)
    data[col_series == value].Stress_Level.value_counts().sort_index().plot.pie(autopct='%1.1f%%', colors=['#D3D3D3','#A9A9A9','#808080', '#F08080', '#DC143C', '#B22222'])
    plt.title(value)

  plt.tight_layout()
  plt.show()
```

Построим круговые диаграммы для всех категориальных переменных

```python
pie_by_stress(data[cat_columns[0]])
```

Здесь не прослеживается явной зависимости уровня стресса от качества сна

```python
pie_by_stress(data[cat_columns[1]])
```
